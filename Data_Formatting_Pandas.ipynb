# Data Formatting using Pandas in Python

```python
# First step is to import the csv file containing the data using pandas and saving it to a dataframe
```

```python
import pandas as pd
df = pd.read_csv('Persons.csv')
df.head(10)
```

```python
# The data above shows that it has some formatting issues like in Salary_range column Medium is written as M, Medium and Med
# The values of M and Med should be replaced by Medium as other values in the column are fully pronounced, e.g. High and Low
```

```python
# Second step would be to get this column of the dataframe
df['Salary_Range']
```

```python
# It returns a series. Lets create a list and replace the other variants of Medium
medium_list = ['M','Med']
# The above list contains the other variants of the Medium
```

```python
# Next step would be to replace these variants by 'Medium' and then make changes to the dataframe
df['Salary_Range'] = df['Salary_Range'].replace(medium_list,'Medium')
```

```python
# Finally, lets call the head method to verify the changes
df.head(10)
```

```python
# This dataframe can now be written to a new csv as
df.to_csv('Person_Clean.csv',sep=',')
```
